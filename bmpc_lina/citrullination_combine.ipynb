### Reproduce Wassims Dataset

```python
import os
import pandas as pd
```

```python
root_dir = '/cmnfs/data/proteomics/ProteomeTools/HCD/citrullination/'

# Create an empty list to store DataFrames
dfs = []
```

```python
for subdir, dirs, files in os.walk(root_dir):
    for file in files:
        # Check if the file is named 'msms.txt'
        if file == 'msms.txt':
            # Construct the full file path
            file_path = os.path.join(subdir, file)
            try:
                # Try reading the file into a DataFrame and append to the list
                df = pd.read_table(file_path)  # Adjust sep based on your file's delimiter
                dfs.append(df)
            except Exception as e:
                # If there's an error, print a message and skip the file
                print(f"Failed to read {file_path}: {e}")

```

```python
len(dfs)
```

```python
# Combine all successfully read DataFrames into one
if dfs:  # Only combine if there are DataFrames in the list
    combined_df = pd.concat(dfs, ignore_index=True)
    # Display the combined DataFrame
    print(combined_df)
else:
    print("No valid data found to combine.")
```

```python
# combined_df.to_csv("~/data/citrullination_all/msms.txt", sep = "\t")
```

```python
import matplotlib.pyplot as plt
```

```python
result = combined_df.groupby('Raw file')['Reverse'].apply(lambda x: (x == '+').sum())

plt.figure(figsize=(8, 6))
plt.hist(result, bins=max(result), edgecolor='black', alpha=0.7)
plt.xlabel('Count of Decoys')
plt.ylabel('Number of Raw files')
plt.title('Histogram of Decoys per Raw File')
plt.show()
```

```python
result[result==20]
```

```python
# filter for score > 70 and only 5 spectra per (peptide, charge, collision energy) group
combined_df_filtered = combined_df[combined_df["Score"] > 70]
```

```python
result_df = combined_df_filtered.groupby(['Sequence', 'Charge']).head(5).reset_index(drop=True)
    
```
